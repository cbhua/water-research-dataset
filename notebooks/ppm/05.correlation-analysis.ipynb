```python
import sys; sys.path.append(2 * "../")

# Packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from matplotlib.colors import LinearSegmentedColormap

from src.utils import (
    calculate_correlation_matrix,
    plot_correlation_matrix,
)

# Repository root path
root = 2 * "../"
```

```python
# Feature names
feat_name = [
    "Raw Flow",
    "Raw Temp",
    "Raw EC",
    "Raw TB",
    "Raw Alk",
    "Raw pH",
    "Raw TOC",
    "PACS ppm",
    "Chl Pre",
    "Sed Chl",
    "Sed TB",
    "Chl Mid"
]
```

### Minute Data Correlation Analysis

```python
minute_data = pd.read_csv(root + "data/ppm/ori/minute.csv")
print(f"Minute Data Range: {minute_data['DateTime'][0]} ~ {minute_data['DateTime'][minute_data.shape[0]-1]}")
print(f"Minute Data Shape: {minute_data.shape}")
minute_data.head()
```

```python
corr_matrix = calculate_correlation_matrix(minute_data.drop(columns=["DateTime"]))
print(f"Correlation Matrix Shape: {corr_matrix.shape}")
```

```python
plot_correlation_matrix(
    corr_matrix=corr_matrix, 
    feat_name=feat_name, 
    save_path=root + "figs/ppm/02.corr_matrix_minute_2016_2021.png"
)
```

```python
# SECTION: plot the corr matrix from 2016 to 2019
minute_data_2016_2019 = minute_data[minute_data['DateTime'] < '2020-01-01']
corr_matrix = calculate_correlation_matrix(minute_data_2016_2019.drop(columns=["DateTime"]))
plot_correlation_matrix(
    corr_matrix=corr_matrix, 
    feat_name=feat_name, 
    save_path=root + "figs/ppm/02.corr_matrix_minute_2016_2019.png"
)
```

```python
# SECTION: plot the corr matrix for 2020
minute_data_2020 = minute_data[(minute_data['DateTime'] >= '2020-01-01') & (minute_data['DateTime'] < '2021-01-01')]
corr_matrix = calculate_correlation_matrix(minute_data_2020.drop(columns=["DateTime"]))
plot_correlation_matrix(
    corr_matrix=corr_matrix, 
    feat_name=feat_name, 
    save_path=root + "figs/ppm/02.corr_matrix_minute_2020.png"
)
```

```python
# SECTION: plot the corr matrix from 2016 to 2020
minute_data_2016_2020 = minute_data[minute_data['DateTime'] < '2021-01-01']
corr_matrix = calculate_correlation_matrix(minute_data_2016_2020.drop(columns=["DateTime"]))
plot_correlation_matrix(
    corr_matrix=corr_matrix, 
    feat_name=feat_name, 
    save_path=root + "figs/ppm/02.corr_matrix_minute_2016_2020.png"
)
```
